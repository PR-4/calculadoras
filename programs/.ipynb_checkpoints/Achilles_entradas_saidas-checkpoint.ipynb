# Entrada Aba COT

```python
# Leitura dos dados da Aba COT (poço 158)
das = pd.read_csv("../inputs/AbaCOT/SES158/Densidade_aparente_seca.txt", sep='\s+', skiprows=1,skipfooter=925,
                 names=('Depth(m)','DAS'))
idades = pd.read_csv("../inputs/AbaCOT/SES158/Idades.txt", sep='\s+', skiprows=1,skipfooter=897,
                 names=('Depth(m)','Idade'))
sf = pd.read_csv("../inputs/AbaCOT/SES158/Sand_Fraction.txt", sep='\s+', skiprows=1044, 
                 names=('Depth(m)','SF'))
pp = pd.read_csv("../inputs/AbaCOT/SES158/Produtividade_primaria.txt", sep='\s+', skiprows=1,skipfooter=897,
                 names=('Depth(m)','PP'))
ts = pd.read_csv("../inputs/AbaCOT/SES158/Taxa_de_sedimentacao.txt", sep='\s+', skiprows=1,skipfooter=897,
                 names=('Depth(m)','TS'))
```

```python
das.head()
```

```python
#Armazena entradas em um numpy array
prof1=np.array(das['Depth(m)'])
prof2=np.array(idades['Depth(m)'])
prof3=np.array(sf['Depth(m)'])
prof4=np.array(pp['Depth(m)'])
prof5=np.array(ts['Depth(m)'])

DAS=np.array(das['DAS'])
AGE=np.array(idades['Idade'])
SF=np.array(sf['SF'])
PP=np.array(pp['PP'])
TS=np.array(ts['TS'])
```

```python
print(np.size(prof1),np.size(prof2),np.size(prof3),np.size(prof4),np.size(prof5))
```

### Calculando a taxa de amostragem 

```python
amostragem1=prof1[1]-prof1[0]
amostragem2=prof2[1]-prof2[0]
amostragem3=prof3[1]-prof3[0]
amostragem4=prof4[1]-prof4[0]

print(amostragem1,amostragem2,amostragem3,amostragem4)
```

```python
padrao={'comprimento':15,
            'altura':15
}
#Desenha os plots
figure2 = plm2(5, padrao)
figure2.plot_s(0,DAS,prof1,{'titulo':'Densidade Aparente \n Seca (g\cm³) ','cor':'k','descricao_x':'(a)'})
figure2.plot_s(1,AGE,prof2,{'titulo':'Idade \n (Ma)','cor':'k','descricao_x':'(b)'})
figure2.plot_s(2,SF,prof3,{'titulo':'Fração de Areia \n (%)','cor':'k','descricao_x':'(c)'})
figure2.plot_s(3,PP,prof4,{'titulo':'Produtividade Primária \n (gC/m²/yr)','cor':'k','descricao_x':'(d)'})
figure2.plot_s(4,TS,prof5,{'titulo':'Taxa de Sedimentação \n (cm/ka)','cor':'k','descricao_x':'(e)'})

#figure2.legenda({'ancoragem':(-0.3, 0.09, 3.0, -0.16),'colunas':2,'ordem':[0,1,2,3,4,5,6,7] })

plt.savefig('../images/159ABACOT.pdf', dpi=30, bbox_inches = 'tight', transparent = True)
```
